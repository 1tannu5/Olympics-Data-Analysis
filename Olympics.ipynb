```python
import pandas as pd
summer_df = pd.read_csv('C:\\Users\\stann\\OneDrive\\Documents\\datasets\\olympics\\summer.csv')
winter_df = pd.read_csv('C:\\Users\\stann\\OneDrive\\Documents\\datasets\\olympics\\winter.csv')
summer_df.head(), winter_df.head()
```

```python
import matplotlib.pyplot as plt

summer_medals_by_year = summer_df.groupby('Year').size()
winter_medals_by_year = winter_df.groupby('Year').size()

plt.figure(figsize=(14, 7))

#Summer Olympics
plt.subplot(1, 2, 1)
plt.plot(summer_medals_by_year.index, summer_medals_by_year.values, marker='o')
plt.title('Number of Medals Over the Years (Summer Olympics)')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.grid(True)

# Winter Olympics
plt.subplot(1, 2, 2)
plt.plot(winter_medals_by_year.index, winter_medals_by_year.values, marker='o', color='orange')
plt.title('Number of Medals Over the Years (Winter Olympics)')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
top_summer_countries = summer_df['Country'].value_counts().head(10)
top_winter_countries = winter_df['Country'].value_counts().head(10)

plt.figure(figsize=(14, 7))

# Summer Olympics
plt.subplot(1, 2, 1)
top_summer_countries.plot(kind='bar', color='skyblue')
plt.title('Top Countries by Medal Count (Summer Olympics)')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Winter Olympics
plt.subplot(1, 2, 2)
top_winter_countries.plot(kind='bar', color='lightcoral')
plt.title('Top Countries by Medal Count (Winter Olympics)')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
summer_medals_by_gender = summer_df['Gender'].value_counts()
winter_medals_by_gender = winter_df['Gender'].value_counts()

plt.figure(figsize=(14, 7))

# Summer Olympics
plt.subplot(1, 2, 1)
summer_medals_by_gender.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightpink'], startangle=140)
plt.title('Distribution of Medals by Gender (Summer Olympics)')
plt.ylabel('')

# Winter Olympics
plt.subplot(1, 2, 2)
winter_medals_by_gender.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightpink'], startangle=140)
plt.title('Distribution of Medals by Gender (Winter Olympics)')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
top_summer_sports = summer_df['Sport'].value_counts().head(10)
top_winter_sports = winter_df['Sport'].value_counts().head(10)

plt.figure(figsize=(14, 7))

# Summer Olympics
plt.subplot(1, 2, 1)
top_summer_sports.plot(kind='bar', color='darkred')
plt.title('Top Sport by Medal Count (Summer Olympics)')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Winter Olympics
plt.subplot(1, 2, 2)
top_winter_sports.plot(kind='bar', color='darkmagenta')
plt.title('Top Sport by Medal Count (Winter Olympics)')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
